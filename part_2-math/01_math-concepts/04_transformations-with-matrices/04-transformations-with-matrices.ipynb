# Computing Transformations with Matrices

## Intro

We saw in the previous chapter that:
> any linear transformation in 3D can be specified by just three vectors, or nine numbers in total.

That was the result of the linear transformations preserving linear combinations, and therefore concluding that:
+ as any vector can be specified as a linear combination of the vectors of the standard basis $ \vec{e_1} $, $ \vec{e_2} $, and $ \vec{e_3} $.
+ it is sufficient to know how any given linear transformation $ T $ affects the vectors of the standard basis.

This means that our *Vector Transformation* mental model in which we could apply a given vector to a transformation can be implemented:
+ as a transformation function, as we did in the last chapter, such as `rotate_z_by`
+ as an arrangement of 9 numbers

![Vector Transformations mental model](pics/vector-transformations-mental-model.png)

When arranged in a series of rows and columns, the numbers that tell us how to execute a linear transformation are called a *matrix*, and they are great computational tools.

## Representing Linear Transformations with Matrices

Suppose $ A $ is a linear transformation in 3D, and we know that $ A(\vec{e_1}) = (1, 1, 1) $, $ A(\vec{e_2}) = (1, 0, -1) $, and $ A(\vec{e_3}) = (0, 1, 1) $.

These three vectors having nine components in total contain all of the information required to specify the linear transformation $ A $.

We'll adopt a new notation called *matrix notation* to work with these numbers.

### Writing Vectors and Linear Transformations as Matrices

Matrices are rectangular grids of numbers. Their particular shapes (number of rows and columns) tell us how we should interpret them.

For example, we can interpret a matrix that is a single columng of numbers as a vector with its entries being the coordinates, ordered top to bottom. In this form, the vectors are called *column vectors*.

For example, the standard basis for 3D can be written as column vectors like this:

$
\vec{e_1} = \begin{bmatrix}
  1 \\
  0 \\
  0
\end{bmatrix} \; , \;
\vec{e_2} = \begin{bmatrix}
  0 \\
  1 \\
  0
\end{bmatrix} \; , \;
\vec{e_3} = \begin{bmatrix}
  0 \\
  1 \\
  0
\end{bmatrix}
$

| NOTE: |
| :---- |
| We're using brackets to denote column vectors, but we could have used parentheses as well. |

Similarly, we can use the same notation to specify $ A(\vec{e_1}) $, $ A(\vec{e_2}) $, $ A(\vec{e_3}) $:

$
A(\vec{e_1}) = \begin{bmatrix}
  1 \\
  1 \\
  1
\end{bmatrix} \; , \;
A(\vec{e_2}) = \begin{bmatrix}
  1 \\
  0 \\
  -1
\end{bmatrix} \; , \;
A(\vec{e_3}) = \begin{bmatrix}
  0 \\
  1 \\
  1
\end{bmatrix}
$

We can go one step further, and create a 3x3 grid representation that concatenates these three column vectors side by side:

$
A = \begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1
\end{pmatrix}
$

| NOTE: |
| :---- |
| We used parentheses to denote the matrix, but we could have used brackets as well. |

We can use the same approach for linear transformations in the plane. In that case, a vector consists of two entries, and therefore 4 numbers are sufficient to specify such transformation:

$
A(\vec{e_1}) = \begin{bmatrix}
  2 \\
  0
\end{bmatrix} \; , \;
A(\vec{e_2}) = \begin{bmatrix}
  0 \\
  2
\end{bmatrix}
$

And the corresponding matrix:

$
D = \begin{pmatrix}
2 & 0 \\
0 & 2 \\ 
\end{pmatrix}
$

### Multiplying a Matrix with a Vector

Now we need to look at the opposite problem: given its matrix, how can we evaluate the linear transformation for any vector.

If a linear transformation $ B $ is represented as a matrix, and a vector $ \vec{v} $ is also represented as a matrix (column vector), we have all the information required to calculate $ B(\vec{v}) $, because:

$
B = \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix}
$, and $
v = \begin{bmatrix}
3 \\
-2 \\
5 \\
\end{bmatrix}
$

And we know that the columns of the matrix represent the application of the transformation to the vectors of the standard basis:

$ 
B(\vec{e_1}) = (0, 0, 1) \\
B(\vec{e_2}) = (2, 1, 0) \\
B(\vec{e_3}) = (1, 0, -1) \\
$

And we know that we can write $ \vec{v} $ as a linear combination of the vectors of the standard basis:

$
\vec{v} = 3 \cdot \vec{e_1} + -2 \cdot \vec{e_2} + 5 \cdot \vec{e_3}
$

And finally, we know that linear transformations preserve linear combinations, so that:

$
B(\vec{v}) = 3 \cdot B(\vec{e_1}) + -2 \cdot B(\vec{e_2}) + 5 \cdot B(vec{e_3})
$

Expanding this using the column vector notation, we can write:

$
B(\vec{v}) = 3 \cdot \begin{pmatrix}
0 \\
0 \\
1 \\
\end{pmatrix} -2 \cdot \begin{pmatrix}
2 \\
1 \\
0 \\
\end{pmatrix} + 5 \cdot \begin{pmatrix}
1 \\
0 \\
-1 \\
\end{pmatrix} = \begin{pmatrix}
0 \\
0 \\
3 \\
\end{pmatrix} + \begin{pmatrix}
-4 \\
-2 \\
0 \\
\end{pmatrix} + \begin{pmatrix}
5 \\
0 \\
-5 \\
\end{pmatrix} = \begin{pmatrix}
1 \\
-2 \\
-2 \\
\end{pmatrix}
$

That is, the result of the linear transformation $ B(\vec{v}) $ is $ (1, -2, -2) $.

Treating the a square matrix as a function that operates on a column vectors is a special case of an operation called *matrix multiplication*.

Written in this form& it is written as:

$
B \vec{v} = \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix} \begin{pmatrix}
3 \\
-2 \\
5 \\
\end{pmatrix} = \begin{pmatrix}
1 \\
-2 \\
-2 \\
\end{pmatrix}
$


As opposed to regular numbers & the order matters when you multiply matrices by vectors. In this case $ B \vec{v} $ is a valid product, but $ \vec{v} B $ is not.

We can write Python code to automate the computation of the result of $ B \vec{v} $.

We can write the matrix as a tuple of tuples (or bidimensional list) and the vector as a regular Python as we've been doing

````python
B = (
  (0, 2, 1),
  (0, 1, 0),
  (1, 0, -1)
)

v = (3, -2, 5)
````

We can get the columns of $ B $ as regular vectors using a trick involving the `zip` function:

````python
B = (
  (0, 2, 1),
  (0, 1, 0),
  (1, 0, -1)
)

assert list(zip((0, 2, 1), (0, 1, 0), (1, 0, -1))) == [(0, 0, 1), (2, 1, 0), (1, 0 , -1)]
assert list(zip(*B)) == [(0, 0, 1), (2, 1, 0), (1, 0, -1)]
````

And we've already developed a `linear_combination()` function, which was defined as:

````python
from vec3d.math import add, scale

scalars = [1, 2, 3]
vectors = [(1, 1, 1), (2, 2, 2), (3, 3, 3)]

def linear_combination(scalars, *vectors):
    return add(*[scale(s, v) for s, v in list(zip(scalars, vectors))])

assert linear_combination(scalars, *vectors) == (14, 14, 14)
````

With these building blocks in place we can write a `multiply_matrix_vector` function, because the result of the multiplication will be the linear combination of the vector components by the column vectors of the matrix:

````python
from vec3d.math import add, scale

def linear_combination(scalars, *vectors):
    return add(*[scale(s, v) for s, v in list(zip(scalars, vectors))])

def multiply_matrix_vector(matrix, vector):
    return linear_combination(vector, *zip(*matrix))

B = (
  (0, 2, 1),
  (0, 1, 0),
  (1, 0, -1)
)

v = (3, -2, 5)

assert multiply_matrix_vector(B, v) == (1, -2, -2)
````

Note:

The `linear_combination` function was added to `vec3d` as of v0.2.0.

As a result, the previous snippet can be simplified when using that version:

````python
from vec3d.math import linear_combination

def multiply_matrix_vector(matrix, vector):
    return linear_combination(vector, *zip(*matrix))

B = (
  (0, 2, 1),
  (0, 1, 0),
  (1, 0, -1)
)

v = (3, -2, 5)

assert multiply_matrix_vector(B, v) == (1, -2, -2)
````

We can write the generic recipe for the multiplication of a matrix by a vector:

$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix} = x \cdot \begin{pmatrix}
a \\
d \\
g \\
\end{pmatrix} + y \cdot \begin{pmatrix}
b \\
e \\
h \\
\end{pmatrix} + z \cdot \begin{pmatrix}
c \\
f \\
i \\
\end{pmatrix} = \begin{pmatrix}
ax + by + cz \\
dx + ey + fz \\
gx + hy + iz \\
\end{pmatrix}
$

A similar formula can be written for the multiplication of a 2x2 matrix by a column vector in the 2D plane:

$
\begin{pmatrix}
a & b \\
c & d \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
\end{pmatrix} = x \cdot \begin{pmatrix}
a \\
c \\
\end{pmatrix} + y \cdot \begin{pmatrix}
b \\
d \\
\end{pmatrix} = \begin{pmatrix}
ax + by \\
cx + dy \\
\end{pmatrix}
$

The mnemonic you can use to remember how to multiply a matrix by a vector is:
+ the first coordinate of the resulting vector is the scalar product of the first row by the vector
+ the second coordinate is the scalar product of the 2nd row by the vector
...

that is:

$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix} = \begin{pmatrix}
(a, b, c) \cdot (x, y, z) \\
(d, e, f) \cdot (x, y, z) \\
(g, h, i) \cdot (x, y, z) \\
\end{pmatrix} = \begin{pmatrix}
ax + by + cz \\
dx + ey + fz \\
gx + hy + iz \\
\end{pmatrix}
$

### Composing Linear Transformations by Matrix Multiplication

We saw in the previous chapter that the composition of any number of linear transformations is also a linear transformation.

Because any linear transformation can be represented as a matrix, any two composed linear transformation should as well be represented by a matrix.

Note that this would be much more efficient than sequentially applying one composition after another, especially if the number of linear transformations to apply is large. By contrast, when using matrices you will just need to know the nine numbers that define the composition of all the transformations to be applied.

Consider two linear transformations $ A $ and $ B $, that we want to apply to a vector $ A(B(\vec{v})) $. The matrices for those two transformations are:

$
A = \begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix}
$, $
B = \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix}
$

We know that to calculate $ A(B(\vec{v})) $, we first need to calculate $ B(\vec{v}) $, and to the resulting vector, we will apply $ A $.

We can write $ A(B(\vec{v})) $ as $ A B \vec{v} $ and therefore write:

$
A(B(\vec{v})) = A B \vec{v} = \begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix}
$

But know, we need to define what should be the result of multiplying the two matrices, which will give us the specification of the matrix for the composition of $ A B $.

$
\begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix} = \begin{pmatrix}
? & ? & ? \\
? & ? & ? \\
? & ? & ? \\
\end{pmatrix}
$

However, we know that the first column of $ B $ is the result of $ B(\vec{e_1}) $, and that the first column of the resulting matrix should be $ A(B(\vec{e_1})) $.

But we also know from the previous section, that is $ B(\vec{e_1}) = \vec{w} $, then $ A(B(\vec{e_1})) = A(\vec{w}) $ can be written as the multiplication of the matrix by the column vector:

$
A w = \begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 \\
0 \\
1 \\
\end{pmatrix} = \begin{pmatrix}
0 \\
1 \\
1 \\
\end{pmatrix}
$

And therefore, we have our first column:

$
\begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix} = \begin{pmatrix}
0 & ? & ? \\
1 & ? & ? \\
1 & ? & ? \\
\end{pmatrix}
$

Similarly, we know that the second column of $ B $ is $ B(\vec{e_2}) $. Therefore $ A(B(\vec{e_2})) = A B(\vec{e_2}) $, and that multiplication will give us the second column of the resulting matrix. And the same we can apply to the third column, to obtain the resulting matrix, that we can write as $ A B $:

$
AB = \begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix} = \begin{pmatrix}
0 & 3 & 1 \\
1 & 2 & 0 \\
1 & 1 & 0 \\
\end{pmatrix}
$

We can use mnemonics instead of reasoning through the whole process each time that we do matrix multiplication. Because multiplying a 3x3 matrix by a column vector is the same as doing 3 dot products, multiplying a 3x3 matrix by another 3x3 matrix is nothing more than 9 dot products.

In particular, if we call $ r_{ij} $ the element that we find at the result matrix in column row $ i $ and column $ j $, we can say that:

$
r_{ij} = (a_{i1}, a_{i2}, a_{i3}) \cdot (b_{1j}, b_{2j}, b{3j}) =  a_{i1} b_{1j} + a_{i2} b{2j} + a_{i3} b_{3j}
$

For example, $ r_{23} $ in our previous example will be:

$
r_{23} = (a_{21}, a_{22}, a_{23}) \cdot (b_{13}, b_{23}, b_{33}) =  (1, 0, 1) \cdot (1, 0, -1) = 0 
$

And, $ r_{22} $:

$
r_{22} = (a_{21}, a_{22}, a_{23}) \cdot (b_{12}, b_{22}, b_{23}) =  (1, 0, 1) \cdot (2, 1, 0) = 2 
$

The same reasoning can be applied to 2x2 matrices.

For instance, the result of multiplying these matrices will be:

$
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
\end{pmatrix} \begin{pmatrix}
0 & -1 \\
1 & 0
\end{pmatrix} = \begin{pmatrix}
2 & -1 \\
4 & -3 \\
\end{pmatrix}
$

### Implementing Matrix Multiplication

Extending the dot product trick that we used to implement the `multiply_matrix_vector` function is the simplest way to achieve the `matrix_multiply` function.

Remember that trick relies in `zip(*b)` returning the column vectors for matrix b.

````python
from vec3d.math import dot

A = (
    (1, 1, 0),
    (1, 0, 1),
    (1, -1, 1)
)

B = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

def matrix_multiply(a, b):
    return tuple(tuple(dot(row_v, col_v) for col_v in zip(*b)) for row_v in a)

assert matrix_multiply(A, B) == (
    (0, 3, 1),
    (1, 2, 0),
    (1, 1, 0),
)
````

Similarly, for the 2D example:

````python
from vec3d.math import dot

A = (
    (1, 2),
    (3, 4),
)

B = (
    (0, -1),
    (1, 0),
)

def matrix_multiply(a, b):
    return tuple(tuple(dot(row_v, col_v) for col_v in zip(*b)) for row_v in a)

assert matrix_multiply(A, B) == (
    (2, -1),
    (4, -3),
)
````

### Dissecting the Implementation of `multiply_matrix_vector` and `matrix_multiply`

The implementation of both `multiply_matrix_vector` and `matrix_multiply` have ended up being one-liners that can be difficult to understand.

This section explains step-by-step how to reach to that implementation from the algebraic formula.

#### Understanding `multiply_matrix_vector` Implementation

The reasoning followed to deduce the algebraic formula for the multiplication of a matrix by a vector concluded that:

$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix} = x \cdot \begin{pmatrix}
a \\
d \\
g \\
\end{pmatrix} + y \cdot \begin{pmatrix}
b \\
e \\
h \\
\end{pmatrix} + z \cdot \begin{pmatrix}
c \\
f \\
i \\
\end{pmatrix}
$


As a a result, the challenge is:
> given a bidimensional tuple, extract the column vectors, and then apply the `linear_combination` function using the scalars from the vector $ (x, y, z) $.

Let's consider:

$
A = \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix}
$

and 

$
v = \begin{pmatrix}
3 \\
-2 \\
5 \\
\end{pmatrix}
$

The multiplication should render: $ (1, -2, -2) $.

Let's start by seeing how we can extract the columns from the matrix.

Typically, you would start with a non-Pythonic approach using `for` loops. That will allow you to develop some intuition about the solution:

````python
A = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

num_rows = len(A)
num_cols = len(A[0])

col_vectors = []
for j in range(num_cols):
    col_vector = []
    for i in range(num_rows):
        col_vector.append(A[i][j])
    col_vectors.append(col_vector)

print(col_vectors)
````

So we see that we need to:
+ extract first elements of row 1, row 2, and row 3
+ extract second elements of row 1, row 2, and row 3
+ extract third elements of row 1, row 2, and row 3

That extraction resembles what the `zip` function does.

Remember that the `zip` function can receive any number of iterables and returns another iterable with the first elements, second elements, etc.

Thus:

````python
A = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

tuple(zip(A[0], A[1], A[2]))
````

We're going on the right direction. The last step is to generalize the way in which we're calling `zip` so that we don't have to hardcode `A[0]`, `A[1]`, ...

We can use the star operator `*` for that:

````python
tuple(zip(*A))
````

There we go! 

Now, we just need to wrap everything up by using the linear combination function, which is declared as:

```python
linear_combination(scalars, *vectors)
```

That is, we can directly pass the vector to multiply as the first parameter, and for `*vectors`, we need to use the star-operator again on the piece of code that gets the column vectors.

We're ready to implement the function:

````python
from vec3d.math import linear_combination

A = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

v = (3, -2, 5)

def multiply_matrix_vector(matrix, vector):
    return linear_combination(v, *zip(*matrix))

assert multiply_matrix_vector(A, v) == (1, -2, -2)
````

#### Understanding `matrix_multiply` Implementation

The reasoning followed to deduce the algebraic formula for the multiplication of a matrix by a vector concluded that if $ R $ is that matrix that results from multiplying two matrices $ A $ and $ B $, and identifying each of its elements by its $ (row, col) $ coordinates:

$
r_{ij} = (a_{i1}, a_{i2}, a_{i3}) \cdot (b_{1j}, b_{2j}, b{3j}) =  a_{i1} b_{1j} + a_{i2} b{2j} + a_{i3} b_{3j}
$

So it seems that using the dot product is the best way to find a good implementation.

Let's consider a couple of matrices and the result and develop some intuition about the solution:

$
\begin{pmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
1 & -1 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 & 2 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1 \\
\end{pmatrix}
$


If we *instantiate* the formula above for $ i = 1 $, $ j = 1 $

$
r_{11} = (a_{11}, a_{12}, a_{13}) \cdot (b_{11}, b_{12}, b{13}) = (1, 1, 0) \cdot (0, 0, 1)
$

Thus we need:
+ Extract the rows from the first matrix
+ Extract the columns from the second matrix

If we do the same for $ i = 1 $, $ j = 2 $

$
r_{12} = (a_{11}, a_{12}, a_{13}) \cdot (b_{21}, b_{22}, b{23}) = (1, 1, 0) \cdot (2, 1, 0) = 3
$

See that we reuse the first *row vector* from $ A $ and iterate over the columns of $ B $.

The 2nd challenge we already solved for the `multiply_matrix_vector`. We just need to use the `zip` function and the `*` operator.

````python
B = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

tuple(zip(*B))
````

Extracting the rows from the first matrix is trivial:

````python
A = (
    (1, 1, 0),
    (1, 0, 1),
    (1, -1, 1),
)

[row_v for row_v in A]
````

Now, we're ready to implement the function. Our intuition also tells us that we will need a nested loop so that we iterate over the rows and columns, and then compute the dot product.

Let's see how this will work:

````python
A = (
    (1, 1, 0),
    (1, 0, 1),
    (1, -1, 1),
)

B = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

for i, row_v in enumerate(A):
    for j, col_v in enumerate(zip(*B)):
        print(f"({i}, {j}) {row_v} \cdot {col_v}")
````

Something that we will need to do is returning the results in a bidimensional tuple. We'll need to wrap every row in a tuple, and wrap the result in another tuple. 

We can use comprehensions for that.

````python
from vec3d.math import dot

A = (
    (1, 1, 0),
    (1, 0, 1),
    (1, -1, 1),
)

B = (
    (0, 2, 1),
    (0, 1, 0),
    (1, 0, -1),
)

def matrix_multiply(a, b):
    return tuple(tuple(dot(row_v, col_v) for col_v in zip(*b)) for row_v in a)

assert matrix_multiply(A, B) == (
    (0, 3, 1),
    (1, 2, 0),
    (1, 1, 0),
    )
````

### 3D Animation with Matrix Transformations

To animate a 3D shape, we redraw a transformed version of the original shape in each frame.

This means that if we're using linear transformations, we will need a new matrix for every new frame of the animation.

Because PyGame's built-in clock keeps track of elapsed time, we can generate matrices whose entries depend on time. That is, we can think of the elements of the matrix as a function that takes the current time $ t $ and returns:

$
\begin{pmatrix}
a(t) & b(t) & c(t) \\
d(t) & e(t) & f(t) \\
g(t) & h(t) & i(t) \\
\end{pmatrix}
$

For example, the function `rotate_y_by(t)` which rotates a particular vector as a function of time can be written as:

$
\begin{pmatrix}
cos(t) & 0 & -sin(t) \\
0 & 1 & 0 \\
sin(t) & 0 & cos(t) \\
\end{pmatrix}
$

Let's try that one out in our teapot.

| NOTE: |
| :---- |
| You need to do `ln -s ../../02_mini-projects/10_mat-lib/mat/ mat` before running the example. |

````python
from math import sin, cos
from rendergl import draw_model, load_triangles

def get_rotation_matrix(t):
    seconds = t / 1000
    return (
        (cos(seconds), 0, -sin(seconds)),
        (0, 1, 0),
        (sin(seconds), 0, cos(seconds)),
    )

draw_model(load_triangles(), get_matrix=get_rotation_matrix)
````

And you will see the teapot rotating about the y axis:

### Exercise

Write a function `infer_matrix(n, transformation)` that takes a dimension (like 2 or 3) and a function that is a vector-based linear transformation and returns an n-by-n square matrix which is the matrix representing the linear transformation.

We know that we can create such matrix by calculating the transformation on the vectors of the standard basis.

As a result, we need to:
+ Infer the vectors of the standard basis, using `n`
+ Transpose a matrix, as we need to put the transformed results in a column

Let's start with the first one:
+ We need to be able to find the shape of the standard basis vectors:
  + for n = 2 => (1, 0), (0,1)
  + for n = 3 => (1, 0, 0), (0, 1, 0)...
+ Then we need to return as many vectors as needed (2 for n = 2, 3 for n = 3, etc.)

Let's first set the stage with the facts we know:

If $ A $ is a linear transformation, we know that it preserves linear combinations:

$

We also saw, that a linear transformation $ A $ can be represented by a matrix, whose columns are the result of applying $ A $ to the vectors of the standard basis $ \{ \vec{e_1}, \vec{e_2}, ...\} $.

Now, we need to prove that matrix multiplication preserves sums and scalar multiples. If we do so, we will prove that matrix multiplication is a linear transformation, and therefore, all the matrices represent linear transformations.


Starting on the 2D plane, the multiplication of a matrix by a vector can be written as:

$
\begin{pmatrix}

\end{pmatrix} \begin{pmatrix}

\end{pmatrix} = \begin{pmatrix}

\end{pmatrix}

$

$
\begin{pmatrix}

\end{pmatrix} \begin{pmatrix}

s_1 u_y + s_2 v_y \\

\end{pmatrix}

A (s_1 \vec{u} + s_2 \vec{v}) = s_1 A \vec{u} + s_2 A \vec{v}
$

Thus multiplying a vector by any 3x3 matrix is a linear transformation of 3D vectors.

### Viewing Square and Non-Square Matrices as Vector Functions

We can think of a 2x2 matrix as the data required to do a given linear transformation of a 2D vector:

![Matrix as a transformation](pics/matrix-as-transformation.png)

Under the hood, the *machine* takes in a 2D vector, performs the matrix multiplication:

$
\begin{pmatrix}
1 & 2 \\
0 & 1 \\
\end{pmatrix} \begin{pmatrix}
1 \\
1 \\
\end{pmatrix}
$

and returns the result: 

$ \begin{pmatrix}
3 \\
1 \\
\end{pmatrix}
$.

We've also discussed that matrix multiplication has certain requirements on the *vectors* it can take. For instance, for our 2x2 matrix, it requires a vector with the shape 2x1 to produce a 2x1 result.

Now, let's consider a non-square matrix such as:

$
\begin{pmatrix}
-2 & -1 & -1 \\
2 & -2 & 1 \\
\end{pmatrix}
$

If we're going to multiply this 2x3 matrix by a column vector, it needs to have the shape 3x1 (that is, a 3D vector) and it will produce a 2x1 vector (that is, a 2D vector).

For example:

$
\begin{pmatrix}
-2 & -1 & -1 \\
2 & -2 & 1 \\
\end{pmatrix} \begin{pmatrix}
0 \\
-1 \\
1 \\
\end{pmatrix} = \begin{pmatrix}
0 \\
3 \\
\end{pmatrix}
$

In general, an m-by-n matrix defines a function taking n-dimensional vectors as inputs and returning m-dimensional vectors as outputs.

Any such function is linear, in the sense that it preserves vector sums and scalar multiples.

But it's not a transformation because it doesn't modify the input &mdash; it returns a vector living in a different number of dimensions.

Because of that, the function represented by an m-by-n matrix is called **linear function** or **linear map**.

### Projection as a Linear Map from 3D to 2D

We already saw a vector function that accepted 3D vectors and produced a 2D vector: the projection of a 3D vector on the x,y plane.

Let's call this function $ P $. Then $ P $ does:

$
P((x, y, z)) = (x, y)
$

Let's write $ P $ as a matrix. Because it accepts 3D vectors, and returns 2D vectors it needs to be a 2x3 matrix.

Let's follow the same reasoning we used for describing transformation as matrix and apply $ P $ to the vectors of the standard basis.

$
P(\vec{e_1}) = P((1, 0, 0)) = (1, 0) \\
P(\vec{e_2}) = P((0, 1, 0)) = (0, 1) \\
P(\vec{e_3}) = P((0, 0, 1)) = (0, 0) \\
$

Let's then put those results as column vectors as we did for 2x2 and 3x3 matrices:

$
P = \begin{pmatrix}
1 & 0 & 0 \\ 
0 & 1 & 0 \\
\end{pmatrix}
$

If we multiply $ P $ by a generic 3D vector $ (a, b, c) $:

$
\begin{pmatrix}
1 & 0 & 0 \\ 
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
a \\
b \\
c \\
\end{pmatrix} = \begin{pmatrix}
a \\
b \\
\end{pmatrix}
$

Thus, it seems that we can use the approach we used for describing linear transformations using square matrices for linear maps described by non-square matrices.

#### Projection Preserves Vector Sums

In order for $ P $ to be linear, we need to prove that:

$
P(s_1 \vec{u} + s_2 \vec{v}) = s_1 P(\vec{u}) + s_2 P(\vec{v})
$

But, $ P(s_1 \vec{u} + s_2 \vec{v}) $ is the matrix multiplication:

$
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} (s_1 \begin{pmatrix}
u_x \\
u_y \\
u_z \\
\end{pmatrix} + s_2 \begin{pmatrix}
v_x \\
v_y \\
v_z \\
\end{pmatrix}) \\
= \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
s_1 u_x + s_2 v_x \\
s_1 u_y + s_2 v_y \\
s_1 u_z + s_2 v_z \\
\end{pmatrix} \\
\begin{pmatrix}
s_1 u_x + s_2 v_x \\
s_1 u_y + s_2 v_y \\
\end{pmatrix} = s_1 \begin{pmatrix}
u_x \\
u_y \\
\end{pmatrix} + s_2 \begin{pmatrix}
v_x \\
v_y \\
\end{pmatrix}
$

But because:

$
P(\vec{u}) = P((u_x, u_y, u_z)) = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
u_x \\
u_y \\
u_z \\
\end{pmatrix} = \begin{pmatrix}
u_x \\
u_y \\
\end{pmatrix}
$

And similarly,

$
P(\vec{v}) = P((v_x, v_y, v_z)) = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
v_x \\
v_y \\
v_z \\
\end{pmatrix} = \begin{pmatrix}
v_x \\
v_y \\
\end{pmatrix}
$


We can say:

$
s_1 \begin{pmatrix}
u_x \\
u_y \\
\end{pmatrix} + s_2 \begin{pmatrix}
v_x \\
v_y \\
\end{pmatrix} = s_1 P(\vec{u}) + s_2 P(\vec{v})
$

And therefore,

$
P(s_1 \vec{u} + s_2 \vec{v}) = s_1 P(\vec{u}) + s_2 P(\vec{v})
$

Let's also see that geometrically with an example.

Let's draw a couple of vectors in 3D, and then calculate the vector sum.

### Composing Linear Maps

Matrices store all of the data required to evaluate a *linear function* on their elements.

Also, the shape of the matrices dictate the dimensions of the input and output vectors.

We can consider the composition of applying several matrices representing linear maps to vectors. The only requirement is that the output of the first one matches the input dimensions expectations for the second one.

For example:

$
M = \begin{pmatrix}
-1 & -1 & 0 \\
-2 & 1 & 2 \\
1 & 0 & -1 \\
\end{pmatrix}
$

and:

$
P = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix}
$

+ $ M $ (being a 3x3 matrix) requires a 3D vector as input and outputs another 3D vector.
+ $ P $ (being a 2x3 matrix) requires a 3D vector as input and generates 2D vector


As a result, $ P(M(\vec{v})) $ can be calculated, and we know how to apply a linear map to a vector using matrix multiplication:

Consider for example $ \vec{v} = (1, 1, 1) $.

$
M(\vec{v}) = \begin{pmatrix}
-1 & -1 & 0 \\
-2 & 1 & 2 \\
1 & 0 & -1 \\
\end{pmatrix} \begin{pmatrix}
1 \\
1 \\
1 \\
\end{pmatrix} = \begin{pmatrix}
-2 \\
1 \\
0 \\
\end{pmatrix}
$

And therefore, $ P(M(\vec{v})) = P(-2, 1, 0) $, which we can calculate with:


$
P = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
-2 \\
1 \\
0 \\
\end{pmatrix} = \begin{pmatrix}
-2 \\
1 \\
\end{pmatrix}
$

Therefore, $ P(M((1, 1, 1))) = (-2, 1) $.

Thinking of $ P $ and $ M $ as matrices, the composition of $ P $ and $ M $ is written $ PM $ and it turns out to be the matrix product.

That is,

$
PM = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
\end{pmatrix} \begin{pmatrix}
-1 & -1 & 0 \\
-2 & 1 & 2 \\
1 & 0 & -1 \\
\end{pmatrix} =
\begin{pmatrix}
-1 & -1 & 0 \\
-2 & 1 & 2 \\
\end{pmatrix}
$

And therefore, if we calculate the composition for the vector $ \vec{v} = (1, 1, 1) $ we get:

$
PM \vec{v} = \begin{pmatrix}
-1 & -1 & 0 \\
-2 & 1 & 2 \\
\end{pmatrix} \begin{pmatrix}
1 \\
1 \\
1 \\
\end{pmatrix} = \begin{pmatrix}
-2 \\
1 \\
\end{pmatrix}
$

As a result, we can also think of the composition of linear maps (as we did with the composition of linear transformations) in terms of multiplications of matrices.

## Translating Vectors with Matrices

One advantage of using matrices is that we can extend the computations they represent to any numbers of dimensions.

Therefore, they become the preferred way to work with dimensions higher than three, because the human brain can't picture well vectors of those higher dimensions.

We'll explore a computation that requires higher number of dimensions: translating vectors using a matrix.

### Making Plane Translations Linear

We already demonstrated that translations are not linear transformations because they don't preserve linear combinations.

However, let's consider the following approach in which we *upgrade* to a higher dimension.

Consider the example we worked with in [01: Linear Algebra in 2D](../01_linear-algebra-2d/01-linear-algebra-2d-concepts.ipynb) in which we draw a dinosaur:

````python
from vec2d.graph import draw, Polygon, Points

dino_points = [(-1, 0),(-2, 1),(-4, 0),(-5, 1),(-2, 2),(-5, 2),(-5, 3),(-4, 4),(-3, 4),(-2, 5),(-1, 5),(1, 2),(3, 1),(6, 4),(5, 1),(3, -1),(1, -2),(2, -3),(1, -4),(-1, -4),(0, -3)]

draw(
    Polygon(*dino_points),
    Points(*dino_points)
)
````

Because translation is not a linear map, there is no way we can find a 2x2 matrix describing the translation of the dinosaur vectors by say $ (3, 1) $.

However, if we think of the dinosaur as a polygon of the 3D space, it turns out we can find a matrix that formulates the translation.

Let's illustrate the trick by first *upgrading* our dinosaur's points to 3D by giving them a particular z coordinate, for example z = 1. Then we will be able to draw the dinosaur in 3D:

````python
from vec3d.graph import draw3d, Points3D, Polygon3D

dino_points = [(-1, 0),(-2, 1),(-4, 0),(-5, 1),(-2, 2),(-5, 2),(-5, 3),(-4, 4),(-3, 4),(-2, 5),(-1, 5),(1, 2),(3, 1),(6, 4),(5, 1),(3, -1),(1, -2),(2, -3),(1, -4),(-1, -4),(0, -3)]

dino_points_3d = [(x, y, 1) for x, y in dino_points]

draw3d(
    Polygon3D(*dino_points_3d),
    Points3D(*dino_points_3d)
)
````

Consider now the following 3x3 matrix, that is the identity matrix for the 2D plane, and performs 

$
\begin{pmatrix}
1 & 0 & 3 \\
0 & 1 & 1 \\
0 & 0 & 1 \\
\end{pmatrix}
$

When applied to a vector, it will result in:

$
\begin{pmatrix}
1 & 0 & 3 \\
0 & 1 & 1 \\
0 & 0 & 1 \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix} = \begin{pmatrix}
x + 3z \\
y + z \\
z \\
\end{pmatrix}
$

That is, x, y coordinates will be translated, while the z coordinate will be preserved.

If we apply it to our dino, we will get:

````python
from vec3d.graph import draw3d, Points3D, Polygon3D, Colors3D, LineStyles3D
from mat import multiply_matrix_vector

dino_points = [(-1, 0),(-2, 1),(-4, 0),(-5, 1),(-2, 2),(-5, 2),(-5, 3),(-4, 4),(-3, 4),(-2, 5),(-1, 5),(1, 2),(3, 1),(6, 4),(5, 1),(3, -1),(1, -2),(2, -3),(1, -4),(-1, -4),(0, -3)]

dino_points_3d = [(x, y, 1) for x, y in dino_points]

A = (
    (1, 0, 3),
    (0, 1, 1),
    (0, 0, 1),
)

dino_points_3d_translated = [multiply_matrix_vector(A, v) for v in dino_points_3d]

draw3d(
    Polygon3D(*dino_points_3d, linestyle=LineStyles3D.DENSELY_DOTTED),
    Points3D(*dino_points_3d, color=Colors3D.BLUE),
    Polygon3D(*dino_points_3d_translated, color=Colors3D.ORANGE),
    Points3D(*dino_points_3d_translated, color=Colors3D.ORANGE),
)
````

So that matrix effectively translates the dino point by (3, 1). We can then transport the result to 2D by simply dropping the z coordinate (which is a linear map and can also be described by a matrix):

````python
from vec2d.graph import draw, Polygon, Points, Colors, LineStyles
from mat import multiply_matrix_vector

dino_points = [(-1, 0),(-2, 1),(-4, 0),(-5, 1),(-2, 2),(-5, 2),(-5, 3),(-4, 4),(-3, 4),(-2, 5),(-1, 5),(1, 2),(3, 1),(6, 4),(5, 1),(3, -1),(1, -2),(2, -3),(1, -4),(-1, -4),(0, -3)]
dino_points_3d = [(x, y, 1) for x, y in dino_points]

A = (
    (1, 0, 3),
    (0, 1, 1),
    (0, 0, 1),
)

dino_points_3d_translated = [multiply_matrix_vector(A, v) for v in dino_points_3d]

B = (
    (1, 0, 0),
    (0, 1, 0),
)

dino_points_translated = [multiply_matrix_vector(B, v) for v in dino_points_3d_translated]

draw(
    Polygon(*dino_points, linestyle=LineStyles.DENSELY_DOTTED),
    Points(*dino_points, color=Colors.BLUE),
    Polygon(*dino_points_translated, color=Colors.ORANGE),
    Points(*dino_points_translated, color=Colors.ORANGE),
)
````

So we've seen that we've been able to use matrix multiplications to perform a translation.

### Finding a 3D Matrix for a 2D Translation

Let's consider the matrix that produced the translation, which we'll call $ T $:

$
T = \begin{pmatrix}
1 & 0 & 3 \\
0 & 1 & 1 \\
0 & 0 & 1 \\
\end{pmatrix}
$

As with all the linear transformations, the columns of that matrix represents the effect of applying that transformation to the standard basis vectors:

$
T \vec{e_1} = (1, 0, 0) \\
T \vec{e_2} = (0, 1, 0) \\
T \vec{e_3} = (3, 1, 1) \\
$

That is, $ e_1 $ and $ e_2 $ are unaffected, and $ e_3 $ changes only its x- and y- components.


Let's see this geometrically. Consider a point of the dinosaur such as $ (6, 4, 1) $. 

That point can be expressed as a linear combination:

$
(6, 4, 1) = 6 \vec{e_1} + 4 \vec{e_2} + \vec{e_3}
$

And because linear transformations preserves linear combinations:

$
T((6, 4, 1)) = T(6 \vec{e_1} + 4 \vec{e_2} + \vec{e_3}) \\
= 6 T(\vec{e_1}) + 4 T(\vec{e_2}) + T(\vec{e_3}) \\
= 6 \vec{e_1} + 4 \vec{e_2} + T(\vec{e_3}) \\
$

And $ T(\vec{e_3}) = (3, 1, 0) + \vec{e_3} $.

Thus:

$
T((6, 4, 1)) \\
= 6 \vec{e_1} + 4 \vec{e_2} + T(\vec{e_3}) \\
= 6 (1, 0, 0) + 4 (0, 1, 0) + (3, 1, 0) + (0, 0, 1) \\
= (6, 4, 1) + (3, 1, 0)
$

That is, the result is translated 3 units in the x+ and 1 unit in the y+.

We can get to the same result using matrices:

$
T = \begin{pmatrix}
1 & 0 & 3 \\
0 & 1 & 1 \\
0 & 0 & 1 \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
\end{pmatrix} = \begin{pmatrix}
x + 3 \\
y + 1 \\
z
\end{pmatrix}
$

That is, we are translating each vector by $ (3, 1, 0) $.

If we reverse-engineer the procedure, we can conclude that:

If you want to translate a collection of 2D vectors by some vector $ (a, b) $ you have to:
1. Move the 2D vectors into the 3D space assigning a z coordinate $ z = 1 $.
2. Multiply the vectors by the matrix:

    $
    \begin{pmatrix}
    1 & 0 & a \\
    0 & 1 & b \\
    0 & 0 & 1 \\
    \end{pmatrix}
    $

3. Delete the z-coordinate of all the vectors that were upgraded to the 3D space.

### Combining Translation with other Linear Transformations

Let's consider the previous matrix that translated our figure by the vector $ (3, 1) $:

$
\begin{pmatrix}
1 & 0 & 3 \\
0 & 1 & 1 \\
0 & 0 & 1 \\
\end{pmatrix}
$

We want leave the last row vector untouched, so that the shape is not moved out of the plane defined by z = 1. We also want to leave the last column as $ (3, 1, 1) $ to perform the translation.

This leave us with the possibility to change the $ \begin{pmatrix}
1 & 0 \\
0 & 1 \\
\end{pmatrix}
$.

For example, if we use the matrix:

$ \begin{pmatrix}
0 & -1 \\
1 & 0 \\
\end{pmatrix}
$ 

we will obtain a 90 counterclockwise rotation.

Note that this is just the matrix resulting from applying that rotation to $ e_1 = (1, 0) $ and $ e_2 = (0, 1) $.

If we then insert it in our translation matrix we will obtain:

$
\begin{pmatrix}
0 & -1 & 3 \\
1 & 0 & 1 \\
0 & 0 & 1 \\
\end{pmatrix}
$

which will rotate any vector by 90 in a counterclockwise fashion and translate it by $ (3, 1) $.

````python
from vec2d.graph import draw, Polygon, Points, Colors, LineStyles
from mat import multiply_matrix_vector

dino_points = [(-1, 0),(-2, 1),(-4, 0),(-5, 1),(-2, 2),(-5, 2),(-5, 3),(-4, 4),(-3, 4),(-2, 5),(-1, 5),(1, 2),(3, 1),(6, 4),(5, 1),(3, -1),(1, -2),(2, -3),(1, -4),(-1, -4),(0, -3)]
dino_points_3d = [(x, y, 1) for x, y in dino_points]

rotate_and_translate = (
    (0, -1, 3),
    (1, 0, 1),
    (0, 0, 1),
)

dino_points_3d_transformed = [multiply_matrix_vector(rotate_and_translate, v) for v in dino_points_3d]

flatten_2d = (
    (1, 0, 0),
    (0, 1, 0),
)

dino_points_transformed = [multiply_matrix_vector(flatten_2d, v) for v in dino_points_3d_transformed]

draw(
    Polygon(*dino_points, linestyle=LineStyles.DENSELY_DOTTED),
    Points(*dino_points, color=Colors.BLUE),
    Polygon(*dino_points_transformed, color=Colors.ORANGE),
    Points(*dino_points_transformed, color=Colors.ORANGE),
)
````

This confirms that we can also use matrix to perform transformations of any kind in 2D with a matrix.

### Translating 3D objects in a 4D world

A 4D vector would be an arrow with some length, width, depth, and one other dimension.

When we built the 3D space from 2D plane, we just added a z-coordinate, and we were able to visualize the 3D space by simply stacking an infinite number of 2D planes at different z-coordinates.

````python
from vec3d.graph import draw3d, Polygon3D, Colors3D

def poly(z):
    return [(-3, -3, z), (-3, 3, z), (3, 3, z), (3, -3, z)]


draw3d(
    Polygon3D(*poly(-2), color=Colors3D.ORANGE),
    Polygon3D(*poly(-1), color=Colors3D.PURPLE),
    Polygon3D(*poly(0), color=Colors3D.BLACK),
    Polygon3D(*poly(1), color=Colors3D.BLUE),
    Polygon3D(*poly(2), color=Colors3D.GRAY),
    Polygon3D(*poly(3), color=Colors3D.GREEN),
    Polygon3D(*poly(4), color=Colors3D.CYAN),
    axes=False
)
````

We can think of four dimensions similarly: a collection of 3D spaces that are indexed by some fourth coordinate, e.g., $ t $.

For each specific value of $ t $, for example $ t = 2 $, we would get a 3D space, in the same way that for any specific value of $ z $, such as $ z = 1 $, we get a 2D plane.

Algebraically, we just need to rely on vector math, and think of a 4D vector as a four-tuple numbers.

By doing so, we won't be limited by the *geometric* representation of vectors, and we'll be able to apply the same tricks we applied to translate a 2D vector by upgrading it to 3D, but this time using 4D vectors.

In particular, we will be able to translate any 3D vector by another vector $ (a, b, c) $ applying the following 4x4 matrix:

$
\begin{pmatrix}
1 & 0 & 0 & a \\
0 & 1 & 0 & b \\
0 & 0 & 1 & c \\
0 & 0 & 0 & 1 \\
\end{pmatrix}
$



We can confirm it by simply multiplying the matrix by a generic 3D vector  $ (x, y, z) $ for which we set an additional coordinate $ t = 1 $:

$
\begin{pmatrix}
1 & 0 & 0 & a \\
0 & 1 & 0 & b \\
0 & 0 & 1 & c \\
0 & 0 & 0 & 1 \\
\end{pmatrix} \begin{pmatrix}
x \\
y \\
z \\
1 \\
\end{pmatrix} = \begin{pmatrix}
x + a \\
y + b \\
z + c \\
1 \\
\end{pmatrix}
$

Now, we can package this generic translation as a transformation function, and use our polygon map to draw a translated teapot using our OpenGL rendering backend:

````python
from mat import multiply_matrix_vector
from rendergl import draw_model, polygon_map, load_triangles

def translate_3d(translation_vector):
    def new_fn(v):
        a, b, c = translation_vector
        x, y, z = v
        translation_matrix = (
            (1, 0, 0, a),
            (0, 1, 0, b),
            (0, 0, 1, c),
            (0, 0, 0, 1),
        )
        vector = (x, y, z, 1)
        x_out, y_out, z_out, _ = multiply_matrix_vector(translation_matrix, vector)
        return (x_out, y_out, z_out)
    return new_fn


draw_model(polygon_map(translate_3d((2, 2, -3)), load_triangles()))
````

If everything goes as expected, you will see how the 3D has been translated by $ (2, 2, -3) $.

R(\vec{e_2}) = (\frac{-\sqrt{2}}{2}, \frac{\sqrt{2}}{2})

\end{pmatrix}

Write a function analogous to `translate_3d` called `translate_4d` that uses a 5x5 matric to translate a 4D vector by another 4D vector.

Run an example to show that coordinates are translated.

````python
from mat import multiply_matrix_vector

def translate_4d(translation_vector):
    def new_fn(v):
        a, b, c, d = translation_vector
        x, y, z, t = v
        translation_matrix = (
            (1, 0, 0, 0, a),
            (0, 1, 0, 0, b),
            (0, 0, 1, 0, c),
            (0, 0, 0, 1, d),
            (0, 0, 0, 0, 1),
        )
        vector = (x, y, z, t, 1)
        x_out, y_out, z_out, t_out, _ = multiply_matrix_vector(translation_matrix, vector)
        return (x_out, y_out, z_out, t_out)
    return new_fn


translation_vector = (1, 2, 3, 4)
v = (10, 20, 30, 40)

assert translate_4d(translation_vector)(v) == (11, 22, 33, 44)
````
